### Escribir código con estilo: PEP8

**PEP8** es quizás el PEP más importante, pues nos enseña cómo escribir código Python que, además de ser funcional, siga las pautas y convenciones sugeridas para el lenguaje. 

La sintaxis de Python es permisiva: no demanda una cantidad estricta de espacios o tabulaciones para la indentación, ni impone nombres de variables o funciones, ni nos exige comentar el código. Aunque esto nos da libertad, también dificulta el trabajo en equipo y el mantenimiento a largo plazo de nuestro código. Si adoptamos un estilo común al de otros desarrolladores, seguramente nos entenderemos mejor con ellos y nuestro código será más ordenado y consistente.

PEP8 describe el estilo seguido por los desarrolladores de Python en la redacción de código para la librería estándar. La guía de estilo completa está disponible en la página de [Python PEP8](https://www.python.org/dev/peps/pep-0008/). Algunas de sus directivas notorias son:

- Usar 4 espacios para indentar. Pueden usarse tabulaciones para respetar código antiguo, pero nunca deben mezclarse espacios y tabs en el mismo documento.
- Cada línea debe tener como máximo 79 caracteres. Líneas más largas pueden continuarse debajo con el símbolo '\' o con el uso de paréntesis.
- Múltiples sentencias 'import' deben ir en líneas diferentes. Primero deben importarse los módulos de librería estándar y luego los de terceros.
- Los operadores aritméticos deben rodearse por espacios. Por ejemplo: `a = 2 + 3`
- El signo '=', cuando es usado para pasar valores a los argumentos de una función, no debe rodearse por espacios. Sería correcto escribir, por ejemplo: `def cociente(numerador=0, divisor=1):`

# Descargar datos de la web: `urllib.request`

Cuando necesitamos descargar archivos de internet para trabajarlos en forma local, una alternativa a la descarga manual es la utilización del módulo `urllib.request`. Utilizando los métodos provistos por urllib.request evitamos la descarga manual, lo que facilita la integración de la descarga en un programa con múltiples funciones y la obtención automática de una gran cantidad de archivos.

### Instalación

No es necesario instalar `urllib.request` pues está incorporada a Python 3. En Python 2 deben utilizarse las librerías `urllib` y `urllib2`.

*NOTA*: `urllib.request` se considera código "legacy" o heredado de Python 3. Código de este tipo sólo existe para dar soporte a alguna función antigua, pero podría desaparecer en un futuro más o menos cercano y ser reemplazado por otros módulos (usualmente) mejores. 

Una alternativa a `urllib.request`, más poderosa pero provista por terceros, es la biblioteca [`Requests`](http://docs.python-requests.org/en/master/).

### Ejemplo

Muchas de las bases de datos bioinformáticas adoptan una sintaxis específica para acceder a su información. Por ejemplo, la base de datos PDB permite acceder al archivo de 4GFP en esta dirección:

`http://files.rcsb.org/view/4GFP.pdb`

Conociendo el código identificador del elemento que nos interesa y la sintaxis correcta, podremos construir la dirección de los datos a descargar. Luego, a través del módulo `urlretrieve` de la biblioteca `urllib.request`, podremos descargar los datos almacenados en esa dirección directamente desde nuestro programa:

```python
# Importar módulo urlretrieve de la biblioteca urllib.request
from urllib.request import urlretrieve

# Construir dirección web (URL)
pdb = "4GFP.pdb"
url = "http://files.rcsb.org/view/" + pdb

# Descargar archivo y guardarlo como 4GFP.pdb en la carpeta de trabajo
# sintaxis: urlretrieve(URL,filename)
urlretrieve("http://files.rcsb.org/view/4GFP.pdb", "4GFP.pdb")
```

# Filogenia: `ETE Toolkit`

La librería **[`ETE Toolkit`](http://etetoolkit.org/)** nos permite construir, manipular, analizar y visualizar árboles filogenéticos. Existen muchas librerías orientadas al manejo de filogenias, como el módulo `Bio.Phylo` incluido en [BioPython](http://biopython.org/wiki/Phylo). Entre ellas, `ETE Toolkit` (*Environment for Tree Exploration Toolkit*) provee una gran variedad de clases y métodos útiles, listos para su aplicación. Al tratarse de un *framework* completo para tareas de filogenia, se require bastante lectura y práctica para sacarle el máximo provecho. Sin embargo, la [documentación proporcionada](http://etetoolkit.org/docs/latest/index.html) es detallada y completa, y los resultados que pueden lograrse hacen valer el esfuerzo inicial.

### Instalación

Las instrucciones de instalación de ETE Toolkit bajo distintas metodologías, en Linux y macOS, están disponibles en su [página oficial](http://etetoolkit.org/download/) (aunque están algo desactualizadas).

La manera más fácil y recomendable de instalar `ETE Toolkit` es a través del siguiente comando:

`conda install -c etetoolkit ete3 ete_toolchain`

Se sugiere revisar que la instalación terminó sin errores con el comando:

`ete3 build check`

`ETE Toolkit` también puede instalarse usando pip, aunque es más complicado. Primero deben instalarse ciertas dependencias listadas a continuación:

`sudo apt-get install python-numpy python-pyqt4 python-pyqt5 python-pyqt5.qtsvg python-pyqt5.qtopengl python-lxml python-six`

Luego sí instalamos `ETE Toolkit` usando `pip`:

`pip install --upgrade ete3`

Si vamos a utilizar los métodos `ete-build` y `ete-evol` (y casi con seguridad lo haremos), será necesario instalar herramientas externas:

`ete3 upgrade-external-tools`


### Ejemplo

Un árbol filogenético se construye a partir de un alineamiento múltiple de secuencias. `ETE Toolkit` nos permite visualizar el árbol y el alineamiento lado a lado. Esto nos permite un primer acercamiento visual para explorar las similitudes y diferencias entre las secuencias según su agrupamiento en el árbol.

En este ejemplo (adaptado del [`ETE Toolkit` cookbook](https://github.com/etetoolkit/cookbook/blob/master/recipes/ete_build_basics.ipynb)) usaremos los archivos *NUP62.fasta* y *NUP62.nwk* incluidos en su carpeta de trabajo. Se trata de un alineamiento y el correspondiente árbol filogenético de la [glicoproteína p62](https://www.uniprot.org/uniprot/P37198), que integra el complejo de poro nuclear.

Primero, leemos el árbol a partir de un archivo externo (en formato [*newick*](http://evolution.genetics.washington.edu/phylip/newicktree.html)). El árbol se almacena como un objeto de la clase `PhyloTree`, que contiene atributos y métodos útiles para trabajar con árboles filogenéticos:

```python
# Importamos la clase PhyloTree
from ete3 import PhyloTree
# Leemos el árbol filogenético
tree = PhyloTree("data/NUP62.nwk")
```

Luego, leemos el alineamiento como un archivo externo (en este caso, en formato FASTA), y lo vinculamos con el árbol ya almacenado:

```python
# Leemos el alineamiento
tree.link_to_alignment("data/NUP62.fasta", alg_format="fasta")
```

Finalmente, graficamos lado a lado el árbol y el alineamiento:

```python
# Generamos el gráfico y lo mostramos en pantalla
tree.render("%%inline", h=600, units="px", dpi=300)
# Generamos el gráfico y lo guardamos como un archivo externo
#tree.render("NUP62.png", h=600, units="px", dpi=300)
```

Como nuestro árbol filogenético pertenece a la clase `PhyloTree`, disponemos de otros métodos que nos permiten trabajar con este árbol. Por ejemplo, con un simple comando podemos buscar el ancestro común más reciente entre dos nodos terminales y graficar el árbol resultante:

```python
tree.get_common_ancestor("Phy004W8WJ_FALPE", "Phy004OQ34_STRCA").render("%%inline", layout="basic")
```

Otra función útil de `ETE Toolkit` es permitirnos la comunicación con la base de datos *NCBI Taxonomy*. El método `NCBITaxa` nos permite descargar una copia de esta base de datos y luego utilizarla para convertir TaxID (códigos identificadores de nivel taxonómico) a los nombres de especies correspondientes. Además, nos permite descargar la clasificación taxonómica completa de una especie, recuperar la topología del árbol que conecta a dos especies dadas, etc.

La primera vez que usemos `NCBITaxa` se descargará la base de datos *NCBI Taxonomy* completa (¡atención, ocupa casi 400MB!). La misma quedará guardada por defecto en `~/.etetoolkit/taxa.sqlite` (por ejemplo, en una instalación típica de Linux del usuario *alumno*, esta carpeta sería `/home/alumno/.etetoolkit/taxa.sqlite`).

```python
# Importamos la clase NCBITaxa
from ete3 import NCBITaxa
# Creamos un objeto de esta clase
ncbi = NCBITaxa()
# Si ya teníamos la base de datos, podemos actualizarla
#ncbi.update_taxonomy_database()
```

Vamos a utilizar algunos métodos de la clase `NCBITaxa` para recuperar los TaxID del linaje completo de una cierta especie, de la cual sólo sabemos que posee el TaxID *9606*:

```python
# Recuperamos la lista de TaxID del linaje completo
lineage = ncbi.get_lineage(9606)
# Imprimimos el linaje, con TaxIDs
print(lineage)
```

Ahora convertimos la lista de TaxID en la lista de nombres taxonómicos correspondientes para cada TaxID. Imprimiendo esta lista obtendremos la clasificación taxonómica completa del TaxID *9606*:

```python
# Traducimos TaxID a nombres taxonómicos
names = ncbi.get_taxid_translator(lineage)
# Imprimimos el linaje, con nombres
print([names[taxid] for taxid in lineage])
```

# Machine Learning: `scikit-learn`

El aprendizaje automático o *machine learning* (ML) es una práctica vinculada a la computación y la inteligencia artificial, que se ocupa de desarrollar y aplicar técnicas capaces de aprender a partir de datos observados para predecir el comportamiento de nuevos datos adicionales. Existen numerosos algoritmos y propósitos para hacer ML, pero todos son dependientes de la cantidad y calidad de los datos de aprendizaje: en general, las técnicas de ML mejoran en su predicción a medida que incorporan más y mejores datos.

Tomemos un ejemplo relacionado con una operación aritmética básica. Todas las computadoras son capaces de ejecutar un algoritmo tradicional para sumar dos números, y pueden aplicar los mismos pasos para obtener el resultado de la adición de cualquier par de números. Sin embargo, también podríamos enseñarle a la computadora a sumar sin usar la receta tradicional. Para eso, deberíamos programar algoritmo de ML al que "enseñaríamos" con muchos ejemplos de operaciones y sus resultados correspondientes: "2 + 3 = 5"; "12 + 13 = 25"; "1000 + 10000 = 11000", etc. A través de ML, la computadora sería capaz de reconocer los patrones escondidos en estos pares de entrada y salida de datos, y devolvernos una respuesta al preguntarle por el resultado de otras operaciones no aprendidas antes, como "22 + 23". Que esta respuesta sea "45", "44" o "3,14" dependerá de la cantidad de datos aprendidos, la claridad del patrón subyacente en ellos y el algoritmos de ML aplicado.

El paquete [`scikit-learn`](http://scikit-learn.org/stable/index.html), o simplemente `sklearn`, provee un conjunto de herramientas sencillas para realizar distintas tareas asociadas al *data mining*. Scikit-learn tiene funciones para clasificación de datos discretos y continuos, *clustering* o agrupamiento de datos, selección de parámetros y modelos de comportamiento, etc.

### Ejemplo

Para nuestro ejemplo trabajaremos con un conjunto de datos relativo a las flores *"Iris"*. Es un dataset tradicional en estadística, propuesto por el famoso matemático Ronald Fisher en 1936. Consiste de 50 muestras de cada una de tres especies diferentes de Iris: *Iris versicolor*, *Iris setosa* e *Iris virginica*. Cada una de estas 150 muestras consiste del ancho y el largo (en centímetros) de los sépalos y los pétalos de la flor correspondiente. 



[fuente](https://www.datacamp.com/community/tutorials/machine-learning-in-r)

Utilizando `scikit-learn`, simularemos una situación de estudio donde conocemos las características mencionadas en algunas muestras de las tres *Iris* y deseamos identificar a qué especie pertenecen otras muestras, sobre la base de mediciones en sus pétalos y sépalos. Para este ejemplo, separaremos el dataset en un conjunto de aprendizaje (que representa lo que ya conocemos) y un conjunto de evaluación (lo que queremos clasificar). Realizaremos la clasificación de muestras desconocidas a través de `scikit-learn`, usando el método de "K vecinos más cercanos" (*K-Nearest Neighbors* o KNN) y adaptando [éste tutorial](http://www.pythondiario.com/2018/01/introduccion-al-machine-learning-9-k.html).

En el algoritmo KNN, un objeto será clasificado en la categoría que determinen la mayoría de sus *k* vecinos más cercanos. Dependerá de nosotros definir cuántos vecinos queremos considerar, pero se usan números positivos y en general pequeños. Así, KNN es la forma algorítmica del proverbio "dime con quién andas y te diré quién eres".

Supongamos el siguiente ejemplo ilustrativo, donde hemos tomado los valores de los parámetros X1 y X2 en diez muestras, cinco de las cuales pertenecen a la clase A o amarilla, y cinco a la clase B o violeta. Si aplicáramos KNN para clasificar la nueva muestra "estrella roja", ¿a qué categoría pertenecería?



[fuente](http://www.pythondiario.com/2018/01/introduccion-al-machine-learning-9-k.html)

Conociendo ya los datos, haremos ahora la predicción con el algoritmo KNN. Primero, preparemos nuestros conjuntos de entrenamiento y de evaluación.

```python
# Dividimos nuestros datos en conjuntos de entrenamiento y de prueba
data_train, data_test, target_train, target_test = train_test_split(iris.data, iris.target)
```

Ahora sí, recurrimos a nuestro método KNN para que aprenda a clasificar datos sobre la base de nuestros datos de entrenamiento.

```python
# Declaramos el algoritmo KNN
knn = KNeighborsClassifier(n_neighbors=5)
```

Empecemos con un ejemplo relativamente sencillo, donde usamos KNN para predecir solamente el largo y el ancho de los sépalos.

```python
# Entrenamos el algoritmo KNN con el método 'fit', usando solamente largo y ancho de sépalos
knn.fit(data_train[:, :2], target_train)
```

```python
# Obtenemos la precisión media del algoritmo al clasificar largo y ancho de sépalos en datos de evaluación
knn.score(data_test[:, :2], target_test)
```

Encontramos que el algoritmo es capaz de acertar la clasificación aproximadamente en 3 de 4 casos. ¡No está mal!

Sin embargo, su desempeño tal vez mejoraría si entrenáramos al algoritmo incluyendo también los datos de largo y ancho de los pétalos. Completen los siguientes comandos para saberlo:

```python
# Entrenamos el algoritmo KNN con el método 'fit', usando todos los datos de entrenamiento. ¡COMPLETAR!
knn.fit()
```

```python
# Obtenemos la precisión media del algoritmo al clasificar las muestras usando todos sus datos. ¡COMPLETAR!
knn.score()
```

Hemos visto que podemos clasificar los datos de evaluación usando nuestro algoritmo KNN ya entrenado. Pero, ¿por qué limitarnos a los datos de evaluación que tenemos?

En el último ejemplo, volveremos a entrenar nuestro algoritmo KNN con el largo y ancho de sépalos de nuestros datos de entrenamiento. Sin embargo, esta vez usaremos el KNN entrenado para clasificar todos las posibles combinaciones de largo y ancho de sépalos cercanas al dataset *Iris*, y generaremos una representación visual de esa clasificación.

```python
# Importamos NumPy
import numpy as np

# Extraemos el largo y el ancho de los pétalos de cada muestra en una nueva variable X
X = iris.data[:, :2]
# Extraemos la categoría a la que pertenece cada muestra en una nueva variable
targets = iris.target
# Extraemos los nombres de las categorías en una nueva variable
target_names = iris.target_names

# Entrenamos y aplicamos el método KNN
knn.fit(x_train[:, :2], y_train)

# Definimos los valores máximos y mínimos de largo y ancho de sépalos que nos interesan
x_min, x_max = iris.data[:, 0].min() - 0.5, iris.data[:, 0].max() + 0.5
y_min, y_max = iris.data[:, 1].min() - 0.5, iris.data[:, 1].max() + 0.5

# Definimos los valores individuales de largos y anchos de sépalos en el rango deseado, en incrementos de 0.1 cm
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
# 'np.meshgrid' construye una grilla con forma de array multidimensional

# Usamos KNN para clasificar cada uno de los valores individuales de largo y ancho de sépalos
# 'ravel' convierte array multidimensional en lista unidimensional
# 'np.c_' concatena arrays en columnas consecutivas
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])  

# Definimos una función accesoria que devuelve índices de los elementos en una lista
def label_to_int(labels):
    return [list(set(labels)).index(y_value) for y_value in labels]

# Creamos un array para convertir los nombres de las clases en valores numéricos distintivos
Z = np.array(label_to_int(Z))
# Le damos al array creado la misma forma del array 'xx', con datos de largo de sépalos individuales
Z = Z.reshape(xx.shape)
```

```python
# Importamos método para usar mapa de colores en gráficos
from matplotlib.colors import LinearSegmentedColormap

# Creamos un gráfico vacío de las dimensiones deseadas
plt.figure(2, figsize=(15, 9))
# Dejamos el gráfico listo para sumarle elementos
plt.clf()

# Definimos lista de colores a utilizar
colors = ['navy', 'green', 'red']
# Definimos variable con mapa de colores a utilizar
cm_targets = LinearSegmentedColormap.from_list('cm_targets', colors, N=3)

# OPCIÓN A
# Definimos un gráfico de contornos con regiones asignadas a cada clase, dejando que pyplot determine límites
cp = plt.contourf(xx, yy, Z, alpha=0.3, cmap=cm_targets)
# Incluimos una barra indicando la escala de colores
plt.colorbar(cp)

# OPCIÓN B
# Conociendo los límites de clases, podríamos generar gráficos más definidos:
#levels = [0.0, 0.75, 1.25, 2.00]
#contour = plt.contour(xx, yy, Z, levels, colors='k', alpha=0.3)
#contour_filled = plt.contourf(xx, yy, Z, levels, cmap=cm_targets, alpha=0.3)
#plt.colorbar(contour_filled)

# Incluimos valores individuales de largos y anchos de sépalos del dataset Iris completo
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(iris.data[targets == i, 0], iris.data[targets == i, 1], 
                color=color, lw=2, label=target_name)

    
# Agregamos una leyenda
plt.legend(loc='best', shadow=False, scatterpoints=1)
# Agregamos etiquetas a los ejes
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
# Agregamos título
plt.title("Iris dataset - KNN (k = 5)")
    
# Generamos el gráfico con todos los elementos incluidos, y lo mostramos en pantalla
plt.show()
```
